```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing=True
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
class Deck:
    def __init__(self):
        self.all_cards=[]
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
    def shuffle(self):
        random.shuffle(self.all_cards)
    def pick_card(self):
        return self.all_cards.pop(0)
```

```python
class Chips:
    def __init__(self):
        self.bet=0
        self.total=100
    def win_bet(self):
        self.total+=self.bet
    def lose_bet(self):
        self.total-=self.bet
```

```python
class Hand:
    def __init__(self):
        self.cards=[]
        self.value=0
        self.aces=0
    def add_card(self,card):
        self.cards.append(card)
        self.value+=values[card.rank]
    def adjust_for_aces(self):
        if self.value+11<21:
            pass
        else: self.value-=10
```

```python
def take_bet(chips):
    while True:
        try:
            chips.bet=int(input("\nEnter number of chips for your bet: "))
        except:
            print("Enter integer value!")

        if chips.bet>chips.total:
            print("Sorry! You are exceeding the available chips")
        else:
            break
    
```

```python
def hit(deck,hand):
    hit_card=deck.pick_card()
    if values[hit_card.rank]==11:
        hand.adjust_for_aces()
    hand.add_card(hit_card)
```

```python
def hit_or_stand(deck,hand):
    global playing
    choice='Wrong'
    print("\nYou wanna hit or stand?")
    while choice not in ['h','s']:
        choice=input("Enter H for hit and S for stand: ").lower()
    
        if choice=='h':
            hit(deck,hand)
        elif choice=='s':
            print("Player stands. Dealer's turn")
#             print("\nPlaying = "+playing+"\n")
            playing=False
        else:
            print("Sorry, didn't understand")
            continue
        break
```

```python
def show_some(player,dealer):
    print("\nDealer's hand: ")
    print("One card hidden")
    print(dealer.cards[1])
    
    print("\nPlayer's hand: ")
    for card in player.cards:
        print(card)
    
def show_all(player,dealer):
    print("\nDealer's hand:")
    for card in dealer.cards:
        print(card)
        
    print("\nPlayer's hand: ")
    for card in player.cards:
        print(card)
```
